```python
import pandas as pd

data = pd.read_csv("voeux.csv")

data.head()
```

```python
init = data.columns.values
new = ["time", "name", 31, 28, 26, 24, 21, 16, 15, 11, 10, 3]

assert len(init) == len(new)
for i in range(len(new)):
    data = data.rename(axis="columns", mapper={init[i]: new[i]})
data.columns
```

```python
sujets = new[2:]
scores = {}
for n in sujets:
    s = 0
    for i in data[n]:
        s += i
    scores[n] = s

rank = [] # élts = (num projet, score)

while len(scores) > 0:
    min_value = min(scores.values())
    min_keys = [k for k in scores if scores[k] == min_value]
    for k in min_keys:
        rank.append((k, scores[k]))
        scores.pop(k)

rank
```
